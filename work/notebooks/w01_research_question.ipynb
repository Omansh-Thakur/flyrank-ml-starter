## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

# Load the starter dataset
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Compute 3 key data metrics
total_pages = len(df)
declining_pages = (df["trend_direction"] == "down").sum()
pct_declining = (declining_pages / total_pages) * 100
high_impact_declining = ((df["trend_direction"] == "down") & (df["impressions_90d"] >= 500)).sum()

print("--- Quick Data Audit ---")
print(f"1. Total pages in starter dataset: {total_pages:,}")
print(f"2. Declining pages (trend_direction == 'down'): {declining_pages:,} ({pct_declining:.1f}%)")
print(f"3. High-impact declining pages (Declining & 90d Impressions >= 500): {high_impact_declining:,}")
```
